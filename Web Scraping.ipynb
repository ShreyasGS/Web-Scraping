### Python code to scrape and parse HTML data from a static website

```python
import requests
from bs4 import BeautifulSoup
import pprint


URL = "https://www.monster.com/jobs/search/?q=Software-Developer\
        &where=Australia"
page = requests.get(URL)     #performs an HTTP request to the given URL. 
                             #It retrieves the HTML data that the server sends back and stores that data in a Python object.

pprint.pprint(page.content)
```

```python
soup = BeautifulSoup(page.content, "html.parser")   #Python library for parsing structured data
results = soup.find(id="ResultsContainer")         #find elements by id

print(results.prettify())
```

```python
# Look for Python jobs
python_jobs = results.find_all("h2", string=lambda t: "python" in t.lower())  
for p_job in python_jobs:
    #print(p_job)
    link = p_job.find("a")["href"]   #finds href in the a tag
    print(p_job.text.strip())
    print(f"Apply here: {link}\n") # print("apply here: {}".format(link))

```

```python
# Print out all available jobs from the scraped webpage
job_elems = results.find_all("section", class_="card-content")    #find elements by html classname
for job_elem in job_elems:
    title_elem = job_elem.find("h2", class_="title")
    company_elem = job_elem.find("div", class_="company")
    location_elem = job_elem.find("div", class_="location")
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
```
